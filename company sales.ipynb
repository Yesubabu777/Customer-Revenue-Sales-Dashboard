```python

condition = (ddf['Date'] >= ddf['Start']) & (ddf['Date'] <= ddf['End'])
ddf = ddf[condition]
ddf
```

```python
result_df= pd.concat([ddf, nddf], ignore_index=True)
result_df
```

```python
result_df.head()
```

```python
result_df.columns
```

```python
result_df.columns = result_df.columns.str.lower().str.replace(' ', '_')
```

```python
### Data cleaning & Transformations
result_df['date'] = pd.to_datetime(result_df['date'])
result_df['date_of_birth'] = pd.to_datetime(result_df['date_of_birth'])
result_df['start'] = pd.to_datetime(result_df['start'])
result_df['end'] = pd.to_datetime(result_df['end'])
```

```python
### Creting new columns
```

```python
### Final Amount Paid (after discount):
result_df['final_price'] = result_df['unit_price'] * result_df['quantity'] * (1 - result_df['discount'])
```

```python
### Profit
result_df['profit'] = result_df['final_price'] - result_df['production_cost']*result_df['quantity']
```

```python
### Age
today = pd.Timestamp('now')
result_df['age'] = (today - result_df['date_of_birth']).dt.days // 365
```

```python
### Month and Year for time-based analysis:

result_df['month'] = result_df['date'].dt.month
result_df['year'] = result_df['date'].dt.year
result_df['month_year'] = result_df['date'].dt.to_period('M')
```
